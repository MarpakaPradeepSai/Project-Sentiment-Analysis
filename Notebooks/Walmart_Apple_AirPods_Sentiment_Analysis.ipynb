## **8.1 Distribution of Ratings**

```python
import plotly.express as px
import plotly.graph_objects as go

# Calculate value counts for ratings
rating_counts = df['Rating'].value_counts().sort_index()

# Calculate total entries
total = len(df)

# Calculate percentages
percentages = (rating_counts / total * 100).round(1)

# Create interactive bar chart
fig = go.Figure()

fig.add_trace(go.Bar(
    x=rating_counts.index,
    y=rating_counts.values,
    marker_color=['#440154', '#3b528b', '#21918c', '#5ec962', '#fde725'][:len(rating_counts)],  # viridis colors
    hovertemplate='<b>Rating: %{x}</b><br>Count: %{y}<br>Percentage: %{customdata:.1f}%<extra></extra>',
    customdata=percentages.values,
    text=[f'{p:.1f}%' for p in percentages.values],
    textposition='outside'
))

fig.update_layout(
    title={
        'text': '<b>Distribution of Ratings</b>',
        'x': 0.5,
        'xanchor': 'center',
        'font': dict(size=22)
    },
    xaxis_title='<b>Rating</b>',
    yaxis_title='<b>Number of Reviews</b>',
    width=800,
    height=500,
    showlegend=False,
    yaxis=dict(range=[0, rating_counts.max() * 1.07])
)

fig.show()
```

## **8.2 Average Rating Over Time**

```python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd

# Convert Review_Date to datetime format
df['Review_Date'] = pd.to_datetime(df['Review_Date'], errors='coerce')

# Group by month and calculate the average rating
df_avg_rating = df.groupby(df['Review_Date'].dt.to_period('M'))['Rating'].mean().reset_index()

# Convert period to string for plotting
df_avg_rating['Review_Date'] = df_avg_rating['Review_Date'].astype(str)

# Also get count of reviews per month for tooltip
df_count = df.groupby(df['Review_Date'].dt.to_period('M'))['Rating'].count().reset_index()
df_avg_rating['Review_Count'] = df_count['Rating'].values

# Create interactive line chart
fig = px.line(
    df_avg_rating,
    x='Review_Date',
    y='Rating',
    markers=True,
    title='Average Rating Over Time'
)

# Update hover template for detailed tooltips
fig.update_traces(
    mode='lines+markers',
    marker=dict(size=10, color='#1f77b4'),
    line=dict(width=2, color='#1f77b4'),
    hovertemplate=(
        '<b>Month:</b> %{x}<br>'
        '<b>Average Rating:</b> %{y:.2f}<br>'
        '<b>Number of Reviews:</b> %{customdata:,}<br>'
        '<extra></extra>'
    ),
    customdata=df_avg_rating['Review_Count']
)

fig.update_layout(
    title='Average Rating Over Time',
    title_x=0.5,
    title_font=dict(
        size=27,
        family='Arial Black'
    ),
    xaxis_title=dict(
        text='Date',
        font=dict(size=14, family='Arial Black')
    ),
    yaxis_title=dict(
        text='Average Rating',
        font=dict(size=14, family='Arial Black')
    ),
    width=1600,
    height=600,
    xaxis=dict(tickangle=0),
    hoverlabel=dict(
        bgcolor="white",
        font_size=12,
        font_family="Arial"
    ),
    hovermode='x unified'
)


fig.show()
```

### **Observation :**
- The data reveals a **U-shaped trend**: ratings were volatile in 2019‚Äì2020, plummeted in 2021‚Äì2022 (hitting **1.44 in Oct 2021**), and rebounded strongly in 2023‚Äì2024, peaking at **4.59 in May 2024**. Key insights include **seasonal patterns** (higher year-end ratings, mid-year dips) and critical anomalies like Oct 2021. To sustain growth, prioritize **customer feedback analysis**, **replicating 2024 successes**, and **proactively addressing risks** to maintain the upward trajectory.

```python
# ============================================================
# PART 1: IMPORT LIBRARIES
# ============================================================

import numpy as np
import pandas as pd
import re
import string
import emoji
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from wordsegment import segment, load
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, f1_score
from sklearn.utils.class_weight import compute_class_weight

# Load wordsegment and stopwords
load()
stop_words = set(stopwords.words('english'))


# ============================================================
# PART 2: MODEL BUILDING FUNCTION (MODIFIED TO RETURN MODELS)
# ============================================================

def build_and_evaluate_models(df, df_name, target_col='Sentiment_SIA', test_size=0.2, random_state=42):
    """
    Build and evaluate Logistic Regression, Naive Bayes, and Random Forest
    on the given dataframe.

    Returns:
    --------
    results : dict
        Dictionary containing all model results and trained models
    """

    print("=" * 70)
    print(f"üìä BUILDING MODELS ON: {df_name}")
    print("=" * 70)

    # -----------------------------------------------------------
    # Step 1: Prepare Data
    # -----------------------------------------------------------
    X = df.drop(target_col, axis=1)
    y = df[target_col].map({'Negative': 0, 'Neutral': 1, 'Positive': 2})

    X_train, X_test, y_train, y_test = train_test_split(
        X, y,
        test_size=test_size,
        random_state=random_state,
        stratify=y
    )

    print(f"\nüìÅ Data Split:")
    print(f"   X_train shape: {X_train.shape}")
    print(f"   X_test shape:  {X_test.shape}")

    # Dictionary to store results
    results = {
        'DataFrame': df_name,
        'Models': {}
    }

    # -----------------------------------------------------------
    # Step 2: Logistic Regression
    # -----------------------------------------------------------
    print("\n" + "-" * 50)
    print("üîπ LOGISTIC REGRESSION")
    print("-" * 50)

    lr_model = LogisticRegression(
        solver='liblinear',
        multi_class='ovr',
        random_state=random_state,
        class_weight='balanced'
    )
    lr_model.fit(X_train, y_train)

    lr_train_acc = lr_model.score(X_train, y_train)
    lr_test_acc = lr_model.score(X_test, y_test)
    lr_test_pred = lr_model.predict(X_test)

    print(f"   Train Accuracy: {lr_train_acc:.4f}")
    print(f"   Test Accuracy:  {lr_test_acc:.4f}")
    print(f"\n   Classification Report:\n")
    print(classification_report(y_test, lr_test_pred))

    results['Models']['Logistic Regression'] = {
        'model': lr_model,
        'train_accuracy': lr_train_acc,
        'test_accuracy': lr_test_acc,
        'predictions': lr_test_pred
    }

    # -----------------------------------------------------------
    # Step 3: Multinomial Naive Bayes
    # -----------------------------------------------------------
    print("\n" + "-" * 50)
    print("üîπ MULTINOMIAL NAIVE BAYES")
    print("-" * 50)

    classes = np.unique(y_train)
    class_weights = compute_class_weight(
        class_weight='balanced',
        classes=classes,
        y=y_train
    )
    class_weight_dict = dict(zip(classes, class_weights))
    sample_weight = np.array([class_weight_dict[c] for c in y_train])

    nb_model = MultinomialNB(alpha=1)
    nb_model.fit(X_train, y_train, sample_weight=sample_weight)

    nb_train_acc = nb_model.score(X_train, y_train)
    nb_test_acc = nb_model.score(X_test, y_test)
    nb_test_pred = nb_model.predict(X_test)

    print(f"   Train Accuracy: {nb_train_acc:.4f}")
    print(f"   Test Accuracy:  {nb_test_acc:.4f}")
    print(f"\n   Classification Report:\n")
    print(classification_report(y_test, nb_test_pred))

    results['Models']['Naive Bayes'] = {
        'model': nb_model,
        'train_accuracy': nb_train_acc,
        'test_accuracy': nb_test_acc,
        'predictions': nb_test_pred
    }

    # -----------------------------------------------------------
    # Step 4: Random Forest Classifier
    # -----------------------------------------------------------
    print("\n" + "-" * 50)
    print("üîπ RANDOM FOREST CLASSIFIER")
    print("-" * 50)

    rf_model = RandomForestClassifier(
        n_estimators=200,
        max_depth=None,
        class_weight='balanced',
        n_jobs=-1,
        random_state=random_state
    )
    rf_model.fit(X_train, y_train)

    rf_train_acc = rf_model.score(X_train, y_train)
    rf_test_acc = rf_model.score(X_test, y_test)
    rf_test_pred = rf_model.predict(X_test)

    print(f"   Train Accuracy: {rf_train_acc:.4f}")
    print(f"   Test Accuracy:  {rf_test_acc:.4f}")
    print(f"\n   Classification Report:\n")
    print(classification_report(y_test, rf_test_pred))

    results['Models']['Random Forest'] = {
        'model': rf_model,
        'train_accuracy': rf_train_acc,
        'test_accuracy': rf_test_acc,
        'predictions': rf_test_pred
    }

    print("=" * 70)
    print(f"‚úÖ COMPLETED: {df_name}")
    print("=" * 70, "\n")

    return results


# ============================================================
# PART 3: RUN MODELS ON ALL 3 DATAFRAMES
# ============================================================

# Define all dataframes
dataframes = {
    'CountVectorizer (Unigram)': cv_df,
    'Bag of Words (N-gram)': bow_df,
    'TF-IDF': tfidf_df
}

# Store all results
all_results = {}

# Run models on each dataframe
for df_name, df in dataframes.items():
    all_results[df_name] = build_and_evaluate_models(df, df_name)


# ============================================================
# PART 4: EXTRACT TRAINED MODELS FROM RESULTS
# ============================================================

# Extract Logistic Regression models from all_results
lr_cv = all_results['CountVectorizer (Unigram)']['Models']['Logistic Regression']['model']
lr_bow = all_results['Bag of Words (N-gram)']['Models']['Logistic Regression']['model']
lr_tfidf = all_results['TF-IDF']['Models']['Logistic Regression']['model']

print("‚úÖ Extracted Logistic Regression models:")
print(f"   - lr_cv:    {type(lr_cv)}")
print(f"   - lr_bow:   {type(lr_bow)}")
print(f"   - lr_tfidf: {type(lr_tfidf)}")
```

```python
# ============================================================
# PART 5: VERIFY YOUR VECTORIZERS EXIST
# ============================================================

# Make sure these vectorizers are defined from your feature engineering step
# They should be the same vectorizers used to create cv_df, bow_df, tfidf_df

# Example - Replace with your actual vectorizer variable names:
# cv = CountVectorizer(...)       # Your unigram vectorizer
# bow = CountVectorizer(...)      # Your n-gram vectorizer
# tfidf = TfidfVectorizer(...)    # Your TF-IDF vectorizer

print("\n‚úÖ Verify your vectorizers are defined:")
print(f"   - cv (CountVectorizer):    {type(cv)}")
print(f"   - bow (N-gram Vectorizer): {type(cv_ngram_range)}")
print(f"   - tfidf (TfidfVectorizer): {type(tfidfv_ngram_max_features)}")


# ============================================================
# PART 6: PREPROCESSING FUNCTION
# ============================================================

def preprocess_review(review):
    """Clean and preprocess the input review."""

    review_lower = review.lower()
    review_no_urls = re.sub(r'http\S+|www\S+|https\S+', '', review_lower, flags=re.MULTILINE)
    review_no_emojis = emoji.replace_emoji(review_no_urls, replace='')
    review_no_punctuation = review_no_emojis.translate(str.maketrans('', '', string.punctuation))
    review_no_special = re.sub(r'[^a-zA-Z\s]', '', review_no_punctuation)
    review_tokenized = word_tokenize(review_no_special)
    review_no_stopwords = [word for word in review_tokenized if word not in stop_words]

    lemmatizer = WordNetLemmatizer()
    review_lemmatized = [lemmatizer.lemmatize(word) for word in review_no_stopwords]

    return ' '.join(review_lemmatized)


# ============================================================
# PART 7: PREDICTION FUNCTION
# ============================================================

def predict_sentiment(review, model, vectorizer):
    """Predict the sentiment of the input review."""

    cleaned_review = preprocess_review(review)
    review_vectorized = vectorizer.transform([cleaned_review])
    sentiment = model.predict(review_vectorized)

    sentiment_labels = {0: 'Negative', 1: 'Neutral', 2: 'Positive'}
    return sentiment_labels[sentiment[0]]


# ============================================================
# PART 8: MANUAL REVIEWS DATA
# ============================================================

manual_reviews = [
    # ========================= POSITIVE REVIEWS (100) =========================
    "I absolutely love my Apple AirPods! The sound quality is incredible, and the noise cancellation is perfect for blocking out background noise. The seamless connection with my iPhone is a game-changer. Worth every penny!",

    "The Apple AirPods are fantastic! They fit comfortably and are incredibly easy to pair with my Apple devices. I especially love the clear sound and the battery life is impressive. Definitely recommend!",

    "I've tried many wireless earbuds, but nothing beats the Apple AirPods. The sound is crisp, and the design is sleek and modern. They're perfect for both work and workouts!",

    "These AirPods exceeded my expectations! The audio is sharp, and the connection never drops. I can use them all day without worrying about the battery.",

    "Absolutely amazing! The comfort and sound quality are top-notch. I use them for calls, music, and even gaming‚Äîthey never disappoint.",

    "Best purchase I've made this year! The AirPods connect instantly to my iPhone and iPad. The sound is crystal clear, and I love how lightweight they are.",

    "These are perfect for my daily commute. The battery lasts all day, and the charging case is super convenient. The sound quality is excellent for podcasts and music.",

    "I'm blown away by how well these integrate with my Apple ecosystem. Switching between my MacBook and iPhone is seamless. The audio quality is superb!",

    "Great product! The AirPods are comfortable for long listening sessions, and the sound is balanced and clear. I use them every single day.",

    "Fantastic wireless earbuds! The setup took literally 2 seconds, and they've been flawless ever since. Battery life is way better than I expected.",

    "I love how compact and portable these are. The charging case fits perfectly in my pocket, and the AirPods never fall out during my runs. Highly recommend!",

    "The sound quality is phenomenal for such small earbuds. Bass is punchy, highs are clear, and the connection is rock solid. Very happy with this purchase!",

    "Best earbuds I've ever owned! They're so easy to use with Siri integration, and the call quality is outstanding. People say I sound crystal clear.",

    "I was hesitant about the price, but these are worth it! The convenience of having them automatically connect and pause when I take them out is brilliant.",

    "Amazing sound quality and battery life! I charge them maybe twice a week, and I use them several hours daily. The design is sleek and comfortable.",

    "These AirPods are a game-changer for my workouts. They stay secure, sound great, and the sweat hasn't affected them at all. Love them!",

    "Perfect for video calls and music! The microphone quality is excellent, and everyone on my Zoom calls says they can hear me perfectly. No regrets!",

    "I'm impressed by how quickly they charge. Just 15 minutes gives me hours of listening time. The sound is crisp and the bass is surprisingly good.",

    "Great investment! The AirPods work flawlessly with all my Apple devices. The double-tap feature to skip songs is super convenient.",

    "I use these for everything‚Äîmusic, calls, podcasts, and audiobooks. The comfort is unmatched, and the battery life is fantastic. Couldn't be happier!",

    "The Hey Siri feature is incredibly useful! I can change songs, make calls, and get directions without touching my phone. These are brilliant!",

    "Excellent sound quality and they fit my ears perfectly. I've worn them for 5+ hours straight with no discomfort. The charging case is a lifesaver.",

    "These AirPods are a must-have for anyone in the Apple ecosystem! They switch between my devices automatically, and the sound is always excellent.",

    "I'm so glad I bought these! The connection is instant, the sound is clear and balanced, and they're so comfortable I forget I'm wearing them.",

    "Outstanding product! The battery lasts all day, and the charging case gives me multiple charges. I use them daily for work calls and music.",

    "The audio quality is impressive! Music sounds rich and detailed, and phone calls are crystal clear. The design is minimalist and elegant.",

    "I love these AirPods! They're perfect for my morning jogs‚Äîthey don't fall out, and the sound quality motivates me to keep running. Great purchase!",

    "Best wireless earbuds on the market! The pairing process is effortless, and the battery life exceeds my expectations. Sound quality is top-tier.",

    "These are worth every dollar! The seamless integration with my iPhone and Apple Watch is amazing. I can't imagine going back to wired headphones.",

    "Incredible sound quality and comfort! I wear them all day at work and never experience ear fatigue. The battery life is exceptional too.",

    "I'm extremely satisfied with my AirPods! They connect instantly, sound fantastic, and the charging case is compact and convenient. Highly recommend!",

    "Perfect for traveling! The battery lasts through long flights, and the sound quality is excellent. They block out airplane noise surprisingly well.",

    "These AirPods are amazing for the price I paid! The sound is clear, the bass is good, and they're super comfortable. I use them every day.",

    "Great for phone calls and music! The microphone is excellent, and people always comment on how clear I sound. The battery life is impressive.",

    "I love the convenience of these AirPods! They automatically connect when I open the case, and the sound quality is fantastic. No complaints!",

    "Best audio experience I've had with wireless earbuds! The connection is stable, the sound is rich, and they're incredibly comfortable.",

    "These are perfect! I use them for running, gym sessions, and commuting. They stay in place, sound great, and the battery lasts forever.",

    "Excellent product! The AirPods are easy to use, sound amazing, and integrate perfectly with my iPhone. I couldn't ask for more.",

    "I'm thrilled with these AirPods! The sound quality is superb, the battery life is long, and they're so comfortable I can wear them all day.",

    "Amazing wireless earbuds! The setup was instant, the sound is crystal clear, and they've never disconnected. I'm very impressed!",

    "These AirPods are fantastic for daily use! They're comfortable, sound great, and the charging case is super convenient. Love them!",

    "Great sound quality and battery life! I charge them once a week, and I use them daily for several hours. They're perfect!",

    "I'm so happy with this purchase! The AirPods are comfortable, easy to use, and the sound quality is excellent. Highly recommend!",

    "These are the best earbuds I've ever owned! The connection is flawless, the sound is amazing, and they're incredibly comfortable.",

    "Perfect for my lifestyle! I use them for workouts, commuting, and working from home. The battery life and sound quality are outstanding.",

    "I love how seamlessly these work with my iPhone! The sound is clear and balanced, and the battery lasts all day. Excellent product!",

    "These AirPods are worth the investment! The sound quality is top-notch, they're super comfortable, and the battery life is impressive.",

    "Fantastic earbuds! The pairing is instant, the sound is excellent, and they fit perfectly. I use them every single day and love them!",

    "Amazing product! The sound quality is crisp, the battery lasts forever, and they're so comfortable. I couldn't be happier with my purchase!",

    "These AirPods are perfect! They sound great, fit comfortably, and the battery life is exceptional. I highly recommend them to everyone!",

    "The clarity on calls is outstanding! My colleagues always comment on how professional I sound. The music quality is equally impressive.",

    "I'm a fitness enthusiast, and these AirPods are perfect for my active lifestyle. They stay put during intense workouts and sound amazing!",

    "The battery life is phenomenal! I can go days without charging the case. The sound is rich and immersive. Totally worth it!",

    "I love the compact design! The case fits easily in my pocket, and the AirPods themselves are so lightweight. Sound quality is top-notch!",

    "These have changed my daily routine! I use them for morning runs, work calls, and evening relaxation. The sound is always perfect.",

    "The automatic pause feature when I remove them is so convenient! The sound quality is excellent, and the battery lasts all day long.",

    "I'm impressed by the build quality! They feel premium, sound incredible, and the connection to my devices is instant every time.",

    "Perfect for students! I use them for online classes, studying with music, and group calls. The microphone and speakers are excellent.",

    "The bass response is surprisingly good for such small earbuds! The highs are crisp, and the mids are well-balanced. Love these!",

    "I bought these for my daily commute, and they're perfect! The noise isolation is great, and the battery easily lasts my entire day.",

    "The seamless switching between my Mac, iPad, and iPhone is incredible! No more manual pairing. These are truly smart earbuds!",

    "I use these for audiobooks and podcasts, and the clarity is outstanding! The battery life means I rarely need to charge them.",

    "The comfort level is unmatched! I can wear them for hours without any ear pain. The sound quality is consistently excellent.",

    "These AirPods have exceeded all my expectations! The connection is stable, the sound is rich, and they're perfect for any activity.",

    "I love the minimalist design! They look great, feel comfortable, and the sound quality is exceptional. Best earbuds I've owned!",

    "The quick charging feature is a lifesaver! A few minutes in the case gives me hours of use. The sound is always crystal clear.",

    "Perfect for working from home! The call quality is excellent, and I can wear them comfortably during long meetings. Love them!",

    "The sound quality rivals expensive over-ear headphones! I'm amazed by the audio performance from such compact earbuds.",

    "I'm a music producer, and these AirPods are surprisingly good for casual listening! The sound signature is balanced and pleasant.",

    "The Hey Siri integration is so useful! I can control everything hands-free while driving or cooking. These are incredibly convenient!",

    "I bought these as a gift for my partner, and they absolutely love them! The sound is great, and they're so easy to use.",

    "The durability is impressive! I've had mine for over a year, and they still work perfectly. The sound quality hasn't degraded at all.",

    "These are perfect for video editing! The low latency means audio stays perfectly synced with video. Great for creative work!",

    "I love how they automatically connect when I put them in my ears! The convenience factor alone makes these worth every penny.",

    "The sound isolation is better than I expected! I can focus on my music without being distracted by outside noise. Excellent!",

    "These AirPods have become an essential part of my day! From morning workouts to evening Netflix, they deliver perfect sound every time.",

    "I'm amazed by the range! I can walk around my entire house without losing connection. The battery life is also fantastic!",

    "The call quality is so clear that people don't even realize I'm using earbuds! The microphones filter out background noise perfectly.",

    "I've dropped these so many times, and they still work flawlessly! The build quality is excellent, and the sound is always great.",

    "Perfect for meditation and relaxation! The sound is soothing and clear, and the comfortable fit means I can use them for extended periods.",

    "The double-tap controls are so intuitive! I can easily skip tracks or answer calls without fumbling with my phone. Love this feature!",

    "I use these for gaming, and the audio quality is impressive! The low latency means I can hear every sound effect clearly and on time.",

    "The battery indicator on my iPhone is so helpful! I always know when to charge them. The battery life itself is outstanding though!",

    "These are the most comfortable earbuds I've ever worn! I sometimes forget I'm wearing them. The sound quality is exceptional too!",

    "I love the matte white finish! They look sleek and modern. The sound quality matches the premium appearance perfectly.",

    "Perfect for language learning! The clear audio helps me catch every nuance in pronunciation. These have been invaluable for my studies!",

    "The charging case is brilliantly designed! It's compact, charges quickly, and provides multiple full charges. So convenient!",

    "I'm a runner, and these AirPods stay secure even during intense sprints! The sweat resistance is great, and the sound motivates me!",

    "The spatial audio feature is incredible! Watching movies with these feels immersive. The sound quality is truly next-level!",

    "I bought these on sale, and they're the best value purchase I've made! The quality, comfort, and features are all outstanding!",

    "The automatic ear detection is brilliant! Music pauses when I remove them and resumes when I put them back. So smart!",

    "These AirPods have improved my productivity! I can take calls while working, and the sound quality keeps me focused on tasks.",

    "I love sharing audio with my friend who also has AirPods! We can watch movies together with synchronized sound. So cool!",

    "The sound quality during workouts is motivating! The bass gives energy, and they never fall out no matter how intense my exercises get!",

    "Perfect for virtual meetings! My voice comes through clearly, and I can hear everyone perfectly. These have been essential for remote work!",

    "I'm impressed by how well these work in noisy environments! The sound clarity remains excellent even on busy streets or in cafes.",

    "The battery efficiency is remarkable! Even with heavy daily use, I only need to charge the case twice a week. Absolutely brilliant!",

    "These AirPods are perfect for my long commutes! The comfort and battery life mean I can enjoy music or podcasts the entire journey!",

    "The unboxing experience alone felt premium! Everything about these AirPods screams quality, from the packaging to the satisfying magnetic click of the case. The audio is pristine, and I'm genuinely impressed!",

    "I'm a sleep podcast listener, and these AirPods are a dream! They're so comfortable I can sleep on my side without discomfort, and the auto-pause saves my battery when I drift off. Absolute lifesaver!",


    # ========================= NEUTRAL REVIEWS (100) =========================
    "The Apple AirPods are decent, but I expected a bit more for the price. The sound is okay, but there are other brands out there that offer similar quality for less money.",

    "Overall, the AirPods are pretty good. They are comfortable, but I do wish they had a little more bass. Battery life is fine, but it could be better for long flights.",

    "I'm neutral about the Apple AirPods. They work well for casual listening, but I've had issues with connectivity occasionally. They're okay but not mind-blowing.",

    "The AirPods are fine for everyday use, but I don't find them exceptional. They do the job, but I'm not blown away by the performance or features.",

    "They are average wireless earbuds. Sound is decent, battery life is standard, and they fit okay. Nothing stands out, but they're not bad either.",

    "These AirPods are good but not great. The sound quality is acceptable, but I've heard better from cheaper alternatives. They're convenient though.",

    "They work as expected. The connection is usually stable, and the battery lasts a decent amount of time. I'm satisfied but not overly impressed.",

    "The AirPods are okay for the price. They're comfortable and easy to use, but the sound quality isn't anything special. They get the job done.",

    "I have mixed feelings about these. The convenience is great, but the audio quality is just average. For the price, I expected a bit more.",

    "They're decent earbuds. The battery life is acceptable, and they connect easily to my iPhone. However, the sound could be richer and fuller.",

    "The AirPods are fine for calls and podcasts, but I'm not thrilled with them for music. The bass is lacking, and they feel a bit overpriced.",

    "They're good but not amazing. The fit is comfortable, and the battery lasts long enough for my needs. I just wish the sound was better.",

    "These are standard wireless earbuds. They connect quickly and work well with Apple devices, but the sound quality is nothing to write home about.",

    "I'm moderately satisfied. The AirPods are convenient and comfortable, but the audio isn't as impressive as I'd hoped. They're just okay.",

    "They work well enough. The battery life is decent, and they pair easily with my phone. I just don't find them worth the premium price.",

    "The AirPods are alright. They're easy to use and fairly comfortable, but the sound quality is average at best. I've had better experiences with other brands.",

    "They're functional but not exceptional. The connection is stable most of the time, and the battery is okay. I expected more for this price point.",

    "These are decent earbuds for casual use. The sound is fine, and they're comfortable for short periods. I wouldn't pay full price though.",

    "The AirPods are okay. They fit well and connect easily, but the audio quality doesn't match the hype. They're good but not great.",

    "I'm neither impressed nor disappointed. They work as advertised, but there's nothing particularly special about them. Average product overall.",

    "They're good for calls but mediocre for music. The bass is weak, and the overall sound lacks depth. They're convenient but not amazing.",

    "The AirPods do what they're supposed to do. They're comfortable and easy to pair, but the sound quality is just okay. Nothing remarkable.",

    "They're decent for the Apple ecosystem, but I don't think they justify the price. Sound is average, and battery life is standard.",

    "These are fine for everyday listening. The comfort is good, and they connect quickly, but the audio quality is nothing special.",

    "The AirPods are acceptable. They work well with my iPhone, but I've noticed occasional connectivity hiccups. Sound is decent but not great.",

    "They're okay for casual use. The battery lasts a decent time, and they're fairly comfortable. I just wish the sound was better for the price.",

    "I'm on the fence about these. They're convenient and work well with Apple products, but the sound quality is just average compared to competitors.",

    "These are decent wireless earbuds. They fit comfortably and connect easily, but I don't think they're worth the premium price tag.",

    "The AirPods are alright. Battery life is okay, and they're easy to use. However, the sound quality is underwhelming for such an expensive product.",

    "They're functional and do the job, but I'm not particularly excited about them. Sound is decent, comfort is fine, but nothing stands out.",

    "These are good for phone calls but average for music. The connection is usually stable, but the audio lacks richness and depth.",

    "The AirPods are fine but not exceptional. They're comfortable and convenient, but the sound quality doesn't justify the high price.",

    "They work well enough for daily use. The battery is okay, and they pair quickly with my devices. I just expected more from Apple.",

    "I'm moderately happy with these. They're comfortable and easy to use, but the sound quality is just average. Not worth full price in my opinion.",

    "The AirPods are decent but nothing special. They connect easily and fit comfortably, but the audio quality is mediocre at best.",

    "They're okay for casual listening. The battery life is acceptable, and they're fairly comfortable. I've heard better sound from cheaper alternatives.",

    "These are standard wireless earbuds. They work fine with my iPhone, but the sound quality is just okay. I'm not overly impressed.",

    "The AirPods are good but not great. They're convenient and comfortable, but the audio lacks the punch I was expecting for this price.",

    "They're decent for everyday use. The connection is usually stable, and the battery lasts a reasonable amount of time. Nothing exceptional though.",

    "I have mixed opinions about these. They're easy to use and comfortable, but the sound quality is average. They're okay but overpriced.",

    "The AirPods are fine for basic needs. They connect quickly and are comfortable, but the audio doesn't impress me. Just average overall.",

    "They work as expected but don't exceed expectations. The sound is okay, the battery is decent, and they're comfortable enough.",

    "These are acceptable wireless earbuds. They pair easily with my Apple devices, but the sound quality is nothing to get excited about.",

    "The AirPods are decent but not worth the hype. They're comfortable and connect well, but the audio quality is just average for the price.",

    "They're good for calls and podcasts but mediocre for music. The bass is weak, and the overall sound lacks depth. Convenient but not amazing.",

    "I'm neither thrilled nor disappointed with these. They work fine for everyday use, but the sound quality and battery life are just average.",

    "The AirPods are okay. They fit well and connect easily to my iPhone, but the audio quality doesn't match the premium price tag.",

    "They're functional and do their job, but I don't find them exceptional. Sound is decent, comfort is fine, but nothing really stands out.",

    "These are fine for casual listening. The battery lasts a decent time, and they're comfortable. I just wish the sound was better for what I paid.",

    "The AirPods are alright. They're convenient and easy to use, but the sound quality is average at best. I expected more from Apple honestly.",

    "They're decent for video calls and podcasts. The microphone works well, but the music quality is just okay. Nothing extraordinary.",

    "The AirPods serve their purpose. They're comfortable for short sessions, but the sound quality doesn't impress me for this price point.",

    "They're okay for general use. The battery is acceptable, and they connect fine. I just don't think they offer great value for money.",

    "These are standard earbuds. They work well with my Apple devices, but the audio quality is comparable to much cheaper alternatives.",

    "The AirPods are fine. They're easy to set up and use, but the sound lacks the clarity I was hoping for. Average at best.",

    "They do what they claim to do. The connection is stable, battery is decent, but I'm not wowed by the performance. Just adequate.",

    "I'm undecided about these. They're convenient for the Apple ecosystem, but the sound quality doesn't justify the premium pricing.",

    "The AirPods are acceptable for daily use. Comfort is good, but the audio quality is nothing special. I've heard better for less money.",

    "They're okay earbuds. Battery life is standard, sound is decent, and they fit fine. I just expected more innovation for the price.",

    "These work fine for basic listening. The connection is reliable, but the sound quality is just average. Not bad, but not great either.",

    "The AirPods are alright for casual use. They're comfortable and connect easily, but the audio doesn't stand out from the competition.",

    "They're decent but unexceptional. The battery lasts okay, the sound is acceptable, but there's nothing that makes them special.",

    "I'm neutral on these. They work as advertised, but the sound quality and features don't feel premium despite the price tag.",

    "The AirPods are okay for iPhone users. The integration is good, but the sound quality is mediocre compared to dedicated audio brands.",

    "They're fine for everyday tasks. Calls sound clear enough, and music is acceptable. I just don't find them worth the investment.",

    "These are average wireless earbuds. They connect well and are comfortable, but the audio quality is nothing to get excited about.",

    "The AirPods work as expected. Battery life is okay, sound is decent, but I don't see what makes them worth the premium price.",

    "They're good for the convenience factor, but the sound quality is just average. I've had similar experiences with cheaper earbuds.",

    "I'm moderately satisfied. They're comfortable and easy to use, but the audio performance doesn't match the cost. Okay but overpriced.",

    "The AirPods are fine for Apple users. The pairing is seamless, but the sound quality is average. I expected more for this investment.",

    "They're decent earbuds. The battery is okay, the fit is comfortable, but the sound lacks the depth I was hoping for at this price.",

    "These work well enough. Connection is stable, battery is acceptable, but the audio quality doesn't justify the premium pricing.",

    "The AirPods are okay. They're convenient for Apple devices, but the sound is just average. I don't think they're worth full price.",

    "They're fine for general use. Comfortable fit, easy pairing, but the audio quality is nothing exceptional. Average overall.",

    "I'm not particularly impressed. They work fine, but the sound quality and battery life are standard. Nothing stands out.",

    "The AirPods are acceptable. They do their job, but I don't find them particularly impressive. Average performance for a high price.",

    "They're okay for everyday listening. The convenience is nice, but the sound quality doesn't match cheaper alternatives I've tried.",

    "These are decent but not remarkable. They connect easily and are comfortable, but the audio is just okay for the price point.",

    "The AirPods work fine for basic needs. Sound is acceptable, battery is decent, but I don't think they offer good value for money.",

    "They're alright. The setup is easy, and they're comfortable, but the sound quality is average at best. Nothing special.",

    "I'm neither happy nor unhappy with these. They work as advertised, but there's nothing that makes them stand out from competitors.",

    "The AirPods are fine for casual use. Battery life is okay, sound is decent, but I expected better quality for this price.",

    "They're decent wireless earbuds. Easy to use and comfortable, but the audio quality doesn't impress me. Average overall.",

    "These work adequately. The connection is stable, and they're comfortable, but the sound is just okay. Not worth the premium in my view.",

    "The AirPods are acceptable for daily use. They connect easily and fit well, but the sound quality is mediocre for the cost.",

    "They're okay. Convenient for Apple users, but the audio quality is standard. I don't see what justifies the high price tag.",

    "I'm lukewarm about these. They work fine, but the sound quality and features don't feel special. Average product at a premium price.",

    "The AirPods are fine. They're easy to use and comfortable, but the sound doesn't impress me. Adequate but not outstanding.",

    "They're decent for what they are. Battery is okay, sound is acceptable, but I've had similar experiences with much cheaper options.",

    "These work as expected. Connection is good, fit is comfortable, but the audio quality is just average. Nothing remarkable.",

    "The AirPods are okay for Apple ecosystem users. Pairing is seamless, but the sound quality is standard. Not worth the premium to me.",

    "They're fine for general listening. Comfortable and easy to use, but the audio performance is average. I expected more for this price.",

    "I'm neutral about these AirPods. They work well enough, but the sound quality and battery life are just okay. Nothing special.",

    "The AirPods are acceptable. They connect easily and are fairly comfortable, but the audio quality doesn't justify the cost.",

    "They're okay wireless earbuds. The convenience is nice, but the sound quality is average compared to the competition.",

    "These work fine for everyday use. Battery is decent, sound is okay, but I don't think they're worth the investment.",

    "The AirPods are alright. Easy to set up and use, but the audio quality is nothing special. Average at this price point.",

    "They're decent but not impressive. Comfortable fit, stable connection, but the sound quality is just okay. Nothing stands out.",

    "I'm undecided about these. They work as advertised, but the performance doesn't feel premium despite the price tag.",

    # ========================= NEGATIVE REVIEWS (100) =========================
    "I regret buying the Apple AirPods. They started disconnecting from my iPhone after just a few weeks, and the sound quality was disappointing for such an expensive product.",

    "Very disappointed with my AirPods. They don't stay in my ears during exercise, and the battery drains too quickly. I was expecting much more for the price.",

    "The AirPods are overrated. The sound is mediocre, and they don't block out any noise despite claiming to have noise cancellation. I wouldn't buy them again.",

    "Terrible experience. The connection drops constantly, and the charging case stopped working within a month. Not worth the money at all.",

    "I had high hopes, but these AirPods are a letdown. The sound is tinny, they hurt my ears, and the battery life is shorter than advertised.",

    "Complete waste of money! The left AirPod stopped working after two months, and customer service was unhelpful. Very frustrating experience.",

    "These are awful. They fall out constantly, the connection is unreliable, and the sound quality is poor. I regret this purchase completely.",

    "Very disappointed! The battery life is terrible‚Äîthey barely last 2 hours on a full charge. The sound is also underwhelming for the price.",

    "Don't waste your money on these! They disconnect randomly, the fit is uncomfortable, and the charging case is flimsy. Terrible product.",

    "I'm extremely unhappy with my AirPods. The connection drops every few minutes, and the sound quality is worse than my old $20 earbuds.",

    "These are overpriced junk! The left AirPod died within weeks, and the sound is muffled and tinny. Apple should be ashamed of this product.",

    "Horrible experience! They won't stay in my ears, the battery drains in an hour, and they keep disconnecting from my phone. Total garbage.",

    "I'm very disappointed. The AirPods are uncomfortable, the sound quality is poor, and they stopped charging after a month. Waste of money!",

    "These are the worst earbuds I've ever owned. The connection is terrible, they fall out constantly, and the battery life is pathetic.",

    "Not worth the hype! The sound is mediocre, they don't fit properly, and the battery dies way too quickly. I expected much better from Apple.",

    "Terrible product! The left AirPod stopped working completely after three weeks. The sound was never great to begin with. Very disappointed.",

    "I hate these AirPods! They're uncomfortable, the connection is spotty, and the battery life is awful. I've wasted my money on this trash.",

    "Very poor quality! The charging case broke within two weeks, and the AirPods themselves have terrible sound. I want my money back!",

    "These are garbage! They disconnect constantly, the sound is tinny, and they hurt my ears after 10 minutes. Absolutely terrible purchase.",

    "I'm furious about this purchase! The AirPods stopped charging after a month, and the sound quality was never good. Complete rip-off!",

    "Worst earbuds ever! They fall out during any movement, the battery dies in less than an hour, and the connection drops constantly. Horrible!",

    "Don't buy these! The left AirPod failed after just a few weeks, and the right one sounds terrible. Apple quality has really gone downhill.",

    "These are absolutely terrible! The fit is uncomfortable, the sound is poor, and they disconnect from my phone every few minutes. Waste of money!",

    "I'm extremely disappointed! The battery life is nowhere near advertised, and the sound quality is worse than my cheap wired earbuds.",

    "Horrible product! The AirPods won't stay in my ears, the connection is unreliable, and the charging case stopped working after two weeks.",

    "These are a complete disaster! The sound is awful, they're uncomfortable, and the left one died after a month. I'm so angry about this purchase!",

    "Don't waste your money! The connection drops constantly, the battery drains in minutes, and the sound quality is terrible. Absolutely awful!",

    "I regret buying these! They're uncomfortable, the sound is tinny and flat, and they disconnect from my devices constantly. Very disappointed!",

    "Terrible quality! The charging case broke within weeks, and the AirPods themselves have poor sound and keep falling out. Complete waste!",

    "These are the worst! They don't fit, they don't sound good, and the battery life is pathetic. I expected so much better from Apple.",

    "Awful experience! The left AirPod stopped working after a month, and the sound was never impressive. Customer service was no help either.",

    "I'm so disappointed! These AirPods are uncomfortable, the connection is terrible, and the battery dies way too fast. Not worth it at all!",

    "Complete garbage! They fall out constantly, the sound is mediocre, and the battery life is much shorter than advertised. Don't buy!",

    "These are terrible! The connection drops every few minutes, they hurt my ears, and the sound quality is worse than my old earbuds.",

    "Very poor product! The AirPods stopped charging properly after two weeks, and the sound was never good. I want a refund!",

    "I hate these! They're uncomfortable, the battery dies quickly, and they disconnect constantly. Worst purchase I've made in years!",

    "Don't waste your hard-earned money! The sound is awful, they don't fit right, and the left one died within a month. Terrible!",

    "These are completely worthless! The connection is unreliable, the battery life is pathetic, and the sound quality is disappointing.",

    "Awful earbuds! They fall out during exercise, the sound is tinny, and the charging case is poorly designed. Very unhappy with this!",

    "I'm extremely frustrated! The AirPods disconnect randomly, the battery drains in an hour, and the sound quality is terrible for the price.",

    "These are a joke! The left AirPod failed after three weeks, and the sound was never impressive. Apple really dropped the ball here.",

    "Horrible investment! They're uncomfortable, the connection is spotty, and the battery life is way shorter than promised. Big disappointment!",

    "Don't buy these AirPods! The sound is poor, they don't stay in my ears, and the charging case stopped working after a few weeks.",

    "I'm so angry about this purchase! The AirPods are uncomfortable, the battery dies fast, and they disconnect constantly. Total waste!",

    "These are the worst earbuds I've tried! The sound is muffled, they fall out easily, and the battery life is nowhere near advertised.",

    "Terrible experience! The left AirPod died after a month, the sound is mediocre, and the fit is uncomfortable. I regret buying these!",

    "Complete disappointment! The connection drops constantly, the battery drains too quickly, and the sound quality is poor. Not recommended!",

    "These are awful! They hurt my ears, the sound is tinny, and they disconnect from my phone every few minutes. Horrible product!",

    "I'm very unhappy! The AirPods won't hold a charge, the sound is underwhelming, and they fall out constantly. Waste of money!",

    "Don't believe the hype! These AirPods are uncomfortable, have terrible battery life, and the sound quality is disappointing. Not worth it!",

    "Terrible build quality! The charging case feels cheap, the AirPods rattle inside, and the sound is poor. Very disappointed with Apple!",

    "These are useless! The battery dies within an hour, they disconnect constantly, and the sound quality is awful. Complete waste of money!",

    "I'm so frustrated! The AirPods won't pair properly, the battery life is terrible, and they're uncomfortable to wear. Horrible product!",

    "Worst purchase ever! The left AirPod stopped working in two weeks, and the right one has terrible sound. Don't waste your money!",

    "These are trash! They don't fit my ears, the sound is poor, and the battery drains incredibly fast. I'm so disappointed!",

    "Horrible quality! The connection is unstable, the sound is tinny, and they hurt my ears after wearing them for just minutes.",

    "I regret this purchase so much! The AirPods are uncomfortable, the battery life is awful, and the sound quality is disappointing.",

    "These are terrible earbuds! They fall out constantly, the connection drops frequently, and the sound is worse than cheap alternatives.",

    "Don't buy these! The charging case stopped working after a month, and the sound quality was never good. Complete disappointment!",

    "I'm extremely unhappy! The battery drains too fast, the fit is uncomfortable, and the sound quality is terrible for this price.",

    "Awful product! The AirPods disconnect randomly, the battery life is pathetic, and the sound is muffled and unclear. Not worth it!",

    "These are garbage! The left AirPod died within weeks, the sound is poor, and the charging case is flimsy. Terrible quality!",

    "I'm so disappointed with these! They're uncomfortable, the battery dies quickly, and the connection is unreliable. Complete waste!",

    "Horrible experience! The sound quality is terrible, they won't stay in my ears, and the battery life is much shorter than advertised.",

    "These are the worst! The connection drops constantly, they're uncomfortable, and the sound quality is disappointing. Don't buy!",

    "I hate these AirPods! The battery drains in an hour, they fall out easily, and the sound is tinny. Terrible investment!",

    "Complete trash! The charging case broke after two weeks, the sound is poor, and they disconnect from my phone constantly.",

    "I'm furious! The AirPods stopped working after a month, the sound was always poor, and customer service was unhelpful. Awful!",

    "These are worthless! The fit is terrible, the sound quality is poor, and the battery life is nowhere near what Apple claims.",

    "Don't waste your money! The left AirPod failed within weeks, the sound is mediocre, and they're uncomfortable to wear. Terrible!",

    "Horrible product! They disconnect every few minutes, the battery dies too quickly, and the sound quality is disappointing.",

    "I'm very disappointed! The AirPods are uncomfortable, the connection is unreliable, and the battery life is pathetic. Not worth it!",

    "These are awful! The sound is tinny, they hurt my ears, and the charging case stopped working after a month. Complete waste!",

    "Terrible quality! The AirPods won't stay in my ears, the battery drains fast, and the sound is worse than cheap earbuds.",

    "I regret buying these! The connection drops constantly, the fit is uncomfortable, and the sound quality is terrible for the price.",

    "These are garbage! The left AirPod died after two weeks, the sound is poor, and the battery life is awful. Don't buy!",

    "Horrible experience! They're uncomfortable, the connection is spotty, and the battery drains in less than an hour. Complete disappointment!",

    "I'm so unhappy! The AirPods fall out constantly, the sound is mediocre, and the charging case broke within a month. Terrible!",

    "These are the worst earbuds I've owned! The battery life is pathetic, the sound is tinny, and they disconnect all the time.",

    "Don't buy these AirPods! The quality is terrible, they're uncomfortable, and the battery dies way too quickly. Complete waste!",

    "I'm extremely frustrated! The sound quality is poor, they won't stay in my ears, and the connection drops constantly. Awful!",

    "These are useless! The left AirPod stopped working after three weeks, and the sound was always disappointing. Not worth it!",

    "Horrible product! The battery drains too fast, the fit is uncomfortable, and the sound quality doesn't justify the price at all.",

    "I hate these! They disconnect constantly, the sound is terrible, and the charging case feels cheap. Worst purchase ever!",

    "Complete disaster! The AirPods are uncomfortable, the battery life is awful, and the sound quality is disappointing. Don't buy!",

    "These are trash! The connection is unreliable, they fall out easily, and the sound is worse than my old wired earbuds.",

    "I'm so disappointed! The battery dies in an hour, the fit is terrible, and the sound quality is poor. Complete waste of money!",

    "Awful quality! The left AirPod failed within a month, the sound is tinny, and they're uncomfortable to wear. Not recommended!",

    "Don't waste your money on these! The connection drops constantly, the battery life is pathetic, and the sound is terrible.",

    "I regret this purchase! The AirPods are uncomfortable, the sound quality is poor, and the battery drains way too quickly.",

    "These are horrible! They won't stay in my ears, the connection is spotty, and the sound quality is disappointing for this price.",

    "Terrible product! The charging case broke after two weeks, the battery life is awful, and the sound is mediocre at best.",

    "I'm very unhappy! The AirPods disconnect randomly, the fit is uncomfortable, and the sound quality is terrible. Don't buy!",

    "These are the worst! The battery drains fast, they fall out constantly, and the sound is tinny and poor. Complete waste!",

    "Horrible investment! The left AirPod died after a month, the sound is disappointing, and the connection is unreliable. Awful!",

    "Don't buy these AirPods! The quality is poor, they're uncomfortable, and the battery life is much shorter than advertised.",

    "I'm extremely disappointed! The sound is terrible, they won't stay in my ears, and the charging case stopped working quickly.",

    "These are garbage! The connection drops constantly, the battery life is pathetic, and the sound quality is awful. Not worth it!",

    "Terrible experience! The AirPods are uncomfortable, the sound is poor, and the battery drains way too fast. Complete disappointment!",

    "I hate these! They fall out during exercise, the connection is unreliable, and the sound quality is terrible for the price.",

    "Awful product! The left AirPod failed within weeks, the sound is mediocre, and the battery life is disappointing. Don't buy!"
]


actual_sentiments = ["Positive"] * 100 + ["Neutral"] * 100 + ["Negative"] * 100


# Verification
print(f"Total reviews: {len(manual_reviews)}")
print(f"Total sentiments: {len(actual_sentiments)}")
print(f"Positive: {actual_sentiments.count('Positive')}")
print(f"Neutral: {actual_sentiments.count('Neutral')}")
print(f"Negative: {actual_sentiments.count('Negative')}")


# ============================================================
# PART 9: DEFINE MODELS DICTIONARY (NOW USING EXTRACTED MODELS)
# ============================================================

models_dict = {
    'CountVectorizer (Unigram)': {
        'model': lr_cv,
        'vectorizer': cv
    },
    'Bag of Words (N-gram)': {
        'model': lr_bow,
        'vectorizer': cv_ngram_range
    },
    'TF-IDF': {
        'model': lr_tfidf,
        'vectorizer': tfidfv_ngram_max_features
    }
}


# ============================================================
# PART 10: TEST ALL MODELS FUNCTION
# ============================================================

def test_all_models(manual_reviews, actual_sentiments, models_dict):
    """Test all models on manual reviews and compare performance."""

    all_predictions = {}
    results_summary = []

    for model_name, model_info in models_dict.items():

        print('\n')
        print("=" * 70)
        print(f"üìä TESTING: {model_name}")
        print("=" * 70)

        model = model_info['model']
        vectorizer = model_info['vectorizer']

        predictions = []
        correct = 0

        # Generate predictions for all reviews
        for review in manual_reviews:
            pred = predict_sentiment(review, model, vectorizer)
            predictions.append(pred)

        # Convert sentiments to numerical format for sklearn metrics
        sentiment_map = {'Negative': 0, 'Neutral': 1, 'Positive': 2}
        y_true = [sentiment_map[s] for s in actual_sentiments]
        y_pred = [sentiment_map[s] for s in predictions]

        all_predictions[model_name] = predictions

        # Calculate per-class accuracy
        print("\n" + "-" * 50)
        print("PER-CLASS ACCURACY")
        print("-" * 50)

        per_class_accuracies = {}
        for class_name, class_num in sentiment_map.items():
            # Get indices for this class in true labels
            class_indices = [i for i, label in enumerate(y_true) if label == class_num]
            if class_indices:
                # Count correct predictions for this class
                class_correct = sum(1 for i in class_indices if y_pred[i] == y_true[i])
                class_accuracy = (class_correct / len(class_indices)) * 100
                per_class_accuracies[class_name] = round(class_accuracy, 2)
                print(f"{class_name} Class Accuracy: {class_accuracy:.2f}%")

        # Calculate overall accuracy
        accuracy = sum(1 for i in range(len(y_true)) if y_pred[i] == y_true[i]) / len(y_true) * 100

        # Get classification report
        print("\n" + "-" * 50)
        print("DETAILED CLASSIFICATION REPORT")
        print("-" * 50)

        # Create classification report
        from sklearn.metrics import classification_report

        # Convert numerical labels back to string names for report
        target_names = ['Negative', 'Neutral', 'Positive']
        report = classification_report(y_true, y_pred, target_names=target_names, digits=4)
        print(report)

        # Calculate macro F1
        from sklearn.metrics import f1_score
        macro_f1 = f1_score(y_true, y_pred, average='macro')

        # Store results with per-class accuracies
        results_summary.append({
            'Model': model_name,
            'Negative Accuracy (%)': per_class_accuracies.get('Negative', 0),
            'Neutral Accuracy (%)': per_class_accuracies.get('Neutral', 0),
            'Positive Accuracy (%)': per_class_accuracies.get('Positive', 0),
            'Overall Accuracy (%)': round(accuracy, 2),
            'Macro F1': round(macro_f1, 4)
        })

    return all_predictions, results_summary


# ============================================================
# PART 11: RUN THE TEST
# ============================================================

all_predictions, results_summary = test_all_models(manual_reviews, actual_sentiments, models_dict)
```

```python
def predict_sentiment(review):
    # Define a more detailed and structured prompt for multilingual and emoji handling
    prompt = f"""
    Task: You are an advanced Multilingual Sentiment Analysis model tasked with classifying Apple AirPods reviews from Walmart based on sentiment. Your goal is to determine whether a review is 'Positive', 'Negative', or 'Neutral', considering the tone, content, and underlying emotions of the review, regardless of the language or the use of emojis.

    Instructions:
    1. Analyze the content, tone, and language of the Apple AirPods review.
    2. Classify the sentiment based on the overall experience described in the review:
    - 'Positive' if the review expresses satisfaction or appreciation for the AirPods's features, performance, or usefulness.
    - 'Negative' if the review indicates dissatisfaction, frustration, or issues with the AirPods's performance, features, or experience.
    - 'Neutral' if the review is balanced or lacks strong sentiment (e.g., informative, unclear, or mixed opinions).
    3. Respond with only one of the following labels: 'Positive', 'Negative', or 'Neutral'. Do not provide any additional explanation or details.

    Review: {review}
    Sentiment:"""

    # Tokenize the prompt
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    # Generate the output with max_new_tokens=1 to ensure a short, precise response
    with torch.no_grad():
        outputs = model.generate(**inputs, max_new_tokens=3, do_sample=False)

    # Decode the output and extract the sentiment
    sentiment = tokenizer.decode(outputs[0], skip_special_tokens=True).strip().split()[-1]

    # Ensure the output is one of the expected labels
    if sentiment not in ["Positive", "Negative", "Neutral"]:
        sentiment = "Neutral"  # Default to neutral if the output is unexpected

    return sentiment

# Predict sentiment for each review
df1["Sentiment"] = df1["Review_Text"].apply(predict_sentiment)
df1
```

```python
# ============================================
# CALCULATE CLASS WEIGHTS FOR IMBALANCE HANDLING
# ============================================
class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.array([0, 1, 2]),
    y=train_labels
)
class_weights = torch.tensor(class_weights, dtype=torch.float32)

print("\n" + "="*50)
print("CLASS WEIGHTS (for balanced learning)")
print("="*50)
print(f"  Negative (0): {class_weights[0]:.4f}")
print(f"  Neutral  (1): {class_weights[1]:.4f}")
print(f"  Positive (2): {class_weights[2]:.4f}")
print("\n(Higher weight = more emphasis on minority class)")
```

```python
# Tokenize the text data for both training and validation
train_encodings = tokenizer(
    train_texts,
    truncation=True,
    padding=True,
    max_length=512
)
val_encodings = tokenizer(
    val_texts,
    truncation=True,
    padding=True,
    max_length=512
)

# Convert to HuggingFace Dataset format
train_dataset = Dataset.from_dict({
    'input_ids': train_encodings['input_ids'],
    'attention_mask': train_encodings['attention_mask'],
    'labels': train_labels
})

val_dataset = Dataset.from_dict({
    'input_ids': val_encodings['input_ids'],
    'attention_mask': val_encodings['attention_mask'],
    'labels': val_labels
})

print(f"Train dataset: {train_dataset}")
print(f"Validation dataset: {val_dataset}")
```

## ‚öñÔ∏è **10.3.6 Handling Class Imbalance with Weighted Cross-Entropy Loss**

```python
# ============================================
# CUSTOM TRAINER WITH WEIGHTED LOSS FOR CLASS IMBALANCE
# ============================================

class WeightedLossTrainer(Trainer):
    """
    Custom Trainer that uses weighted CrossEntropyLoss to handle class imbalance.
    Minority classes get higher weights, making the model pay more attention to them.
    """

    def __init__(self, class_weights=None, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.class_weights = class_weights

    def compute_loss(self, model, inputs, return_outputs=False, **kwargs):
        """
        Override the compute_loss method to use weighted CrossEntropyLoss
        """
        labels = inputs.pop("labels")
        outputs = model(**inputs)
        logits = outputs.logits

        # Move class weights to the same device as the model
        if self.class_weights is not None:
            weight = self.class_weights.to(logits.device)
            loss_fct = nn.CrossEntropyLoss(weight=weight)
        else:
            loss_fct = nn.CrossEntropyLoss()

        loss = loss_fct(logits.view(-1, self.model.config.num_labels), labels.view(-1))

        return (loss, outputs) if return_outputs else loss
```

```python
# Define the compute_metrics function with detailed metrics
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = logits.argmax(axis=-1)

    # Overall metrics
    accuracy = accuracy_score(labels, predictions)

    # Macro metrics (treats all classes equally - good for imbalanced data)
    precision_macro = precision_score(labels, predictions, average="macro", zero_division=0)
    recall_macro = recall_score(labels, predictions, average="macro", zero_division=0)
    f1_macro = f1_score(labels, predictions, average="macro", zero_division=0)

    # Weighted metrics (accounts for class imbalance)
    precision_weighted = precision_score(labels, predictions, average="weighted", zero_division=0)
    recall_weighted = recall_score(labels, predictions, average="weighted", zero_division=0)
    f1_weighted = f1_score(labels, predictions, average="weighted", zero_division=0)

    # Per-class F1 scores
    f1_per_class = f1_score(labels, predictions, average=None, zero_division=0)

    return {
        "eval_accuracy": accuracy,
        "eval_precision_macro": precision_macro,
        "eval_recall_macro": recall_macro,
        "eval_f1_macro": f1_macro,
        "eval_precision_weighted": precision_weighted,
        "eval_recall_weighted": recall_weighted,
        "eval_f1_weighted": f1_weighted,
        "eval_f1_negative": f1_per_class[0],
        "eval_f1_neutral": f1_per_class[1],
        "eval_f1_positive": f1_per_class[2],
    }
```

```python
# Set up training arguments with optimizations for imbalanced data
training_args = TrainingArguments(
    output_dir='./results_albert_balanced',
    num_train_epochs=10,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=16,
    warmup_steps=500,
    weight_decay=0.01,
    logging_dir='./logs_albert_balanced',
    logging_steps=10,
    eval_strategy="epoch",
    save_strategy="epoch",
    load_best_model_at_end=True,
    metric_for_best_model="eval_f1_macro",  # Use macro F1 for imbalanced data
    greater_is_better=True,
    save_total_limit=10,
    learning_rate=2e-5,
    lr_scheduler_type="linear"
)
```

##     

## üèãÔ∏è‚Äç‚ôÇÔ∏è **10.3.7 Model Training**

```python
# ============================================
# TRAINER
# ============================================

# OPTION 1: Weighted Loss Trainer (Recommended for moderate imbalance)
trainer = WeightedLossTrainer(
    class_weights=class_weights,
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    compute_metrics=compute_metrics
)

print("Trainer initialized with class-weighted loss!")
print(f"Class weights: {class_weights}")
```

```python
# ============================================
# DEFINE MANUAL TEST REVIEWS
# ============================================

manual_reviews = [
    # ========================= POSITIVE REVIEWS (100) =========================
    "I absolutely love my Apple AirPods! The sound quality is incredible, and the noise cancellation is perfect for blocking out background noise. The seamless connection with my iPhone is a game-changer. Worth every penny!",

    "The Apple AirPods are fantastic! They fit comfortably and are incredibly easy to pair with my Apple devices. I especially love the clear sound and the battery life is impressive. Definitely recommend!",

    "I've tried many wireless earbuds, but nothing beats the Apple AirPods. The sound is crisp, and the design is sleek and modern. They're perfect for both work and workouts!",

    "These AirPods exceeded my expectations! The audio is sharp, and the connection never drops. I can use them all day without worrying about the battery.",

    "Absolutely amazing! The comfort and sound quality are top-notch. I use them for calls, music, and even gaming‚Äîthey never disappoint.",

    "Best purchase I've made this year! The AirPods connect instantly to my iPhone and iPad. The sound is crystal clear, and I love how lightweight they are.",

    "These are perfect for my daily commute. The battery lasts all day, and the charging case is super convenient. The sound quality is excellent for podcasts and music.",

    "I'm blown away by how well these integrate with my Apple ecosystem. Switching between my MacBook and iPhone is seamless. The audio quality is superb!",

    "Great product! The AirPods are comfortable for long listening sessions, and the sound is balanced and clear. I use them every single day.",

    "Fantastic wireless earbuds! The setup took literally 2 seconds, and they've been flawless ever since. Battery life is way better than I expected.",

    "I love how compact and portable these are. The charging case fits perfectly in my pocket, and the AirPods never fall out during my runs. Highly recommend!",

    "The sound quality is phenomenal for such small earbuds. Bass is punchy, highs are clear, and the connection is rock solid. Very happy with this purchase!",

    "Best earbuds I've ever owned! They're so easy to use with Siri integration, and the call quality is outstanding. People say I sound crystal clear.",

    "I was hesitant about the price, but these are worth it! The convenience of having them automatically connect and pause when I take them out is brilliant.",

    "Amazing sound quality and battery life! I charge them maybe twice a week, and I use them several hours daily. The design is sleek and comfortable.",

    "These AirPods are a game-changer for my workouts. They stay secure, sound great, and the sweat hasn't affected them at all. Love them!",

    "Perfect for video calls and music! The microphone quality is excellent, and everyone on my Zoom calls says they can hear me perfectly. No regrets!",

    "I'm impressed by how quickly they charge. Just 15 minutes gives me hours of listening time. The sound is crisp and the bass is surprisingly good.",

    "Great investment! The AirPods work flawlessly with all my Apple devices. The double-tap feature to skip songs is super convenient.",

    "I use these for everything‚Äîmusic, calls, podcasts, and audiobooks. The comfort is unmatched, and the battery life is fantastic. Couldn't be happier!",

    "The Hey Siri feature is incredibly useful! I can change songs, make calls, and get directions without touching my phone. These are brilliant!",

    "Excellent sound quality and they fit my ears perfectly. I've worn them for 5+ hours straight with no discomfort. The charging case is a lifesaver.",

    "These AirPods are a must-have for anyone in the Apple ecosystem! They switch between my devices automatically, and the sound is always excellent.",

    "I'm so glad I bought these! The connection is instant, the sound is clear and balanced, and they're so comfortable I forget I'm wearing them.",

    "Outstanding product! The battery lasts all day, and the charging case gives me multiple charges. I use them daily for work calls and music.",

    "The audio quality is impressive! Music sounds rich and detailed, and phone calls are crystal clear. The design is minimalist and elegant.",

    "I love these AirPods! They're perfect for my morning jogs‚Äîthey don't fall out, and the sound quality motivates me to keep running. Great purchase!",

    "Best wireless earbuds on the market! The pairing process is effortless, and the battery life exceeds my expectations. Sound quality is top-tier.",

    "These are worth every dollar! The seamless integration with my iPhone and Apple Watch is amazing. I can't imagine going back to wired headphones.",

    "Incredible sound quality and comfort! I wear them all day at work and never experience ear fatigue. The battery life is exceptional too.",

    "I'm extremely satisfied with my AirPods! They connect instantly, sound fantastic, and the charging case is compact and convenient. Highly recommend!",

    "Perfect for traveling! The battery lasts through long flights, and the sound quality is excellent. They block out airplane noise surprisingly well.",

    "These AirPods are amazing for the price I paid! The sound is clear, the bass is good, and they're super comfortable. I use them every day.",

    "Great for phone calls and music! The microphone is excellent, and people always comment on how clear I sound. The battery life is impressive.",

    "I love the convenience of these AirPods! They automatically connect when I open the case, and the sound quality is fantastic. No complaints!",

    "Best audio experience I've had with wireless earbuds! The connection is stable, the sound is rich, and they're incredibly comfortable.",

    "These are perfect! I use them for running, gym sessions, and commuting. They stay in place, sound great, and the battery lasts forever.",

    "Excellent product! The AirPods are easy to use, sound amazing, and integrate perfectly with my iPhone. I couldn't ask for more.",

    "I'm thrilled with these AirPods! The sound quality is superb, the battery life is long, and they're so comfortable I can wear them all day.",

    "Amazing wireless earbuds! The setup was instant, the sound is crystal clear, and they've never disconnected. I'm very impressed!",

    "These AirPods are fantastic for daily use! They're comfortable, sound great, and the charging case is super convenient. Love them!",

    "Great sound quality and battery life! I charge them once a week, and I use them daily for several hours. They're perfect!",

    "I'm so happy with this purchase! The AirPods are comfortable, easy to use, and the sound quality is excellent. Highly recommend!",

    "These are the best earbuds I've ever owned! The connection is flawless, the sound is amazing, and they're incredibly comfortable.",

    "Perfect for my lifestyle! I use them for workouts, commuting, and working from home. The battery life and sound quality are outstanding.",

    "I love how seamlessly these work with my iPhone! The sound is clear and balanced, and the battery lasts all day. Excellent product!",

    "These AirPods are worth the investment! The sound quality is top-notch, they're super comfortable, and the battery life is impressive.",

    "Fantastic earbuds! The pairing is instant, the sound is excellent, and they fit perfectly. I use them every single day and love them!",

    "Amazing product! The sound quality is crisp, the battery lasts forever, and they're so comfortable. I couldn't be happier with my purchase!",

    "These AirPods are perfect! They sound great, fit comfortably, and the battery life is exceptional. I highly recommend them to everyone!",

    "The clarity on calls is outstanding! My colleagues always comment on how professional I sound. The music quality is equally impressive.",

    "I'm a fitness enthusiast, and these AirPods are perfect for my active lifestyle. They stay put during intense workouts and sound amazing!",

    "The battery life is phenomenal! I can go days without charging the case. The sound is rich and immersive. Totally worth it!",

    "I love the compact design! The case fits easily in my pocket, and the AirPods themselves are so lightweight. Sound quality is top-notch!",

    "These have changed my daily routine! I use them for morning runs, work calls, and evening relaxation. The sound is always perfect.",

    "The automatic pause feature when I remove them is so convenient! The sound quality is excellent, and the battery lasts all day long.",

    "I'm impressed by the build quality! They feel premium, sound incredible, and the connection to my devices is instant every time.",

    "Perfect for students! I use them for online classes, studying with music, and group calls. The microphone and speakers are excellent.",

    "The bass response is surprisingly good for such small earbuds! The highs are crisp, and the mids are well-balanced. Love these!",

    "I bought these for my daily commute, and they're perfect! The noise isolation is great, and the battery easily lasts my entire day.",

    "The seamless switching between my Mac, iPad, and iPhone is incredible! No more manual pairing. These are truly smart earbuds!",

    "I use these for audiobooks and podcasts, and the clarity is outstanding! The battery life means I rarely need to charge them.",

    "The comfort level is unmatched! I can wear them for hours without any ear pain. The sound quality is consistently excellent.",

    "These AirPods have exceeded all my expectations! The connection is stable, the sound is rich, and they're perfect for any activity.",

    "I love the minimalist design! They look great, feel comfortable, and the sound quality is exceptional. Best earbuds I've owned!",

    "The quick charging feature is a lifesaver! A few minutes in the case gives me hours of use. The sound is always crystal clear.",

    "Perfect for working from home! The call quality is excellent, and I can wear them comfortably during long meetings. Love them!",

    "The sound quality rivals expensive over-ear headphones! I'm amazed by the audio performance from such compact earbuds.",

    "I'm a music producer, and these AirPods are surprisingly good for casual listening! The sound signature is balanced and pleasant.",

    "The Hey Siri integration is so useful! I can control everything hands-free while driving or cooking. These are incredibly convenient!",

    "I bought these as a gift for my partner, and they absolutely love them! The sound is great, and they're so easy to use.",

    "The durability is impressive! I've had mine for over a year, and they still work perfectly. The sound quality hasn't degraded at all.",

    "These are perfect for video editing! The low latency means audio stays perfectly synced with video. Great for creative work!",

    "I love how they automatically connect when I put them in my ears! The convenience factor alone makes these worth every penny.",

    "The sound isolation is better than I expected! I can focus on my music without being distracted by outside noise. Excellent!",

    "These AirPods have become an essential part of my day! From morning workouts to evening Netflix, they deliver perfect sound every time.",

    "I'm amazed by the range! I can walk around my entire house without losing connection. The battery life is also fantastic!",

    "The call quality is so clear that people don't even realize I'm using earbuds! The microphones filter out background noise perfectly.",

    "I've dropped these so many times, and they still work flawlessly! The build quality is excellent, and the sound is always great.",

    "Perfect for meditation and relaxation! The sound is soothing and clear, and the comfortable fit means I can use them for extended periods.",

    "The double-tap controls are so intuitive! I can easily skip tracks or answer calls without fumbling with my phone. Love this feature!",

    "I use these for gaming, and the audio quality is impressive! The low latency means I can hear every sound effect clearly and on time.",

    "The battery indicator on my iPhone is so helpful! I always know when to charge them. The battery life itself is outstanding though!",

    "These are the most comfortable earbuds I've ever worn! I sometimes forget I'm wearing them. The sound quality is exceptional too!",

    "I love the matte white finish! They look sleek and modern. The sound quality matches the premium appearance perfectly.",

    "Perfect for language learning! The clear audio helps me catch every nuance in pronunciation. These have been invaluable for my studies!",

    "The charging case is brilliantly designed! It's compact, charges quickly, and provides multiple full charges. So convenient!",

    "I'm a runner, and these AirPods stay secure even during intense sprints! The sweat resistance is great, and the sound motivates me!",

    "The spatial audio feature is incredible! Watching movies with these feels immersive. The sound quality is truly next-level!",

    "I bought these on sale, and they're the best value purchase I've made! The quality, comfort, and features are all outstanding!",

    "The automatic ear detection is brilliant! Music pauses when I remove them and resumes when I put them back. So smart!",

    "These AirPods have improved my productivity! I can take calls while working, and the sound quality keeps me focused on tasks.",

    "I love sharing audio with my friend who also has AirPods! We can watch movies together with synchronized sound. So cool!",

    "The sound quality during workouts is motivating! The bass gives energy, and they never fall out no matter how intense my exercises get!",

    "Perfect for virtual meetings! My voice comes through clearly, and I can hear everyone perfectly. These have been essential for remote work!",

    "I'm impressed by how well these work in noisy environments! The sound clarity remains excellent even on busy streets or in cafes.",

    "The battery efficiency is remarkable! Even with heavy daily use, I only need to charge the case twice a week. Absolutely brilliant!",

    "These AirPods are perfect for my long commutes! The comfort and battery life mean I can enjoy music or podcasts the entire journey!",

    "The unboxing experience alone felt premium! Everything about these AirPods screams quality, from the packaging to the satisfying magnetic click of the case. The audio is pristine, and I'm genuinely impressed!",

    "I'm a sleep podcast listener, and these AirPods are a dream! They're so comfortable I can sleep on my side without discomfort, and the auto-pause saves my battery when I drift off. Absolute lifesaver!",


    # ========================= NEUTRAL REVIEWS (100) =========================
    "The Apple AirPods are decent, but I expected a bit more for the price. The sound is okay, but there are other brands out there that offer similar quality for less money.",

    "Overall, the AirPods are pretty good. They are comfortable, but I do wish they had a little more bass. Battery life is fine, but it could be better for long flights.",

    "I'm neutral about the Apple AirPods. They work well for casual listening, but I've had issues with connectivity occasionally. They're okay but not mind-blowing.",

    "The AirPods are fine for everyday use, but I don't find them exceptional. They do the job, but I'm not blown away by the performance or features.",

    "They are average wireless earbuds. Sound is decent, battery life is standard, and they fit okay. Nothing stands out, but they're not bad either.",

    "These AirPods are good but not great. The sound quality is acceptable, but I've heard better from cheaper alternatives. They're convenient though.",

    "They work as expected. The connection is usually stable, and the battery lasts a decent amount of time. I'm satisfied but not overly impressed.",

    "The AirPods are okay for the price. They're comfortable and easy to use, but the sound quality isn't anything special. They get the job done.",

    "I have mixed feelings about these. The convenience is great, but the audio quality is just average. For the price, I expected a bit more.",

    "They're decent earbuds. The battery life is acceptable, and they connect easily to my iPhone. However, the sound could be richer and fuller.",

    "The AirPods are fine for calls and podcasts, but I'm not thrilled with them for music. The bass is lacking, and they feel a bit overpriced.",

    "They're good but not amazing. The fit is comfortable, and the battery lasts long enough for my needs. I just wish the sound was better.",

    "These are standard wireless earbuds. They connect quickly and work well with Apple devices, but the sound quality is nothing to write home about.",

    "I'm moderately satisfied. The AirPods are convenient and comfortable, but the audio isn't as impressive as I'd hoped. They're just okay.",

    "They work well enough. The battery life is decent, and they pair easily with my phone. I just don't find them worth the premium price.",

    "The AirPods are alright. They're easy to use and fairly comfortable, but the sound quality is average at best. I've had better experiences with other brands.",

    "They're functional but not exceptional. The connection is stable most of the time, and the battery is okay. I expected more for this price point.",

    "These are decent earbuds for casual use. The sound is fine, and they're comfortable for short periods. I wouldn't pay full price though.",

    "The AirPods are okay. They fit well and connect easily, but the audio quality doesn't match the hype. They're good but not great.",

    "I'm neither impressed nor disappointed. They work as advertised, but there's nothing particularly special about them. Average product overall.",

    "They're good for calls but mediocre for music. The bass is weak, and the overall sound lacks depth. They're convenient but not amazing.",

    "The AirPods do what they're supposed to do. They're comfortable and easy to pair, but the sound quality is just okay. Nothing remarkable.",

    "They're decent for the Apple ecosystem, but I don't think they justify the price. Sound is average, and battery life is standard.",

    "These are fine for everyday listening. The comfort is good, and they connect quickly, but the audio quality is nothing special.",

    "The AirPods are acceptable. They work well with my iPhone, but I've noticed occasional connectivity hiccups. Sound is decent but not great.",

    "They're okay for casual use. The battery lasts a decent time, and they're fairly comfortable. I just wish the sound was better for the price.",

    "I'm on the fence about these. They're convenient and work well with Apple products, but the sound quality is just average compared to competitors.",

    "These are decent wireless earbuds. They fit comfortably and connect easily, but I don't think they're worth the premium price tag.",

    "The AirPods are alright. Battery life is okay, and they're easy to use. However, the sound quality is underwhelming for such an expensive product.",

    "They're functional and do the job, but I'm not particularly excited about them. Sound is decent, comfort is fine, but nothing stands out.",

    "These are good for phone calls but average for music. The connection is usually stable, but the audio lacks richness and depth.",

    "The AirPods are fine but not exceptional. They're comfortable and convenient, but the sound quality doesn't justify the high price.",

    "They work well enough for daily use. The battery is okay, and they pair quickly with my devices. I just expected more from Apple.",

    "I'm moderately happy with these. They're comfortable and easy to use, but the sound quality is just average. Not worth full price in my opinion.",

    "The AirPods are decent but nothing special. They connect easily and fit comfortably, but the audio quality is mediocre at best.",

    "They're okay for casual listening. The battery life is acceptable, and they're fairly comfortable. I've heard better sound from cheaper alternatives.",

    "These are standard wireless earbuds. They work fine with my iPhone, but the sound quality is just okay. I'm not overly impressed.",

    "The AirPods are good but not great. They're convenient and comfortable, but the audio lacks the punch I was expecting for this price.",

    "They're decent for everyday use. The connection is usually stable, and the battery lasts a reasonable amount of time. Nothing exceptional though.",

    "I have mixed opinions about these. They're easy to use and comfortable, but the sound quality is average. They're okay but overpriced.",

    "The AirPods are fine for basic needs. They connect quickly and are comfortable, but the audio doesn't impress me. Just average overall.",

    "They work as expected but don't exceed expectations. The sound is okay, the battery is decent, and they're comfortable enough.",

    "These are acceptable wireless earbuds. They pair easily with my Apple devices, but the sound quality is nothing to get excited about.",

    "The AirPods are decent but not worth the hype. They're comfortable and connect well, but the audio quality is just average for the price.",

    "They're good for calls and podcasts but mediocre for music. The bass is weak, and the overall sound lacks depth. Convenient but not amazing.",

    "I'm neither thrilled nor disappointed with these. They work fine for everyday use, but the sound quality and battery life are just average.",

    "The AirPods are okay. They fit well and connect easily to my iPhone, but the audio quality doesn't match the premium price tag.",

    "They're functional and do their job, but I don't find them exceptional. Sound is decent, comfort is fine, but nothing really stands out.",

    "These are fine for casual listening. The battery lasts a decent time, and they're comfortable. I just wish the sound was better for what I paid.",

    "The AirPods are alright. They're convenient and easy to use, but the sound quality is average at best. I expected more from Apple honestly.",

    "They're decent for video calls and podcasts. The microphone works well, but the music quality is just okay. Nothing extraordinary.",

    "The AirPods serve their purpose. They're comfortable for short sessions, but the sound quality doesn't impress me for this price point.",

    "They're okay for general use. The battery is acceptable, and they connect fine. I just don't think they offer great value for money.",

    "These are standard earbuds. They work well with my Apple devices, but the audio quality is comparable to much cheaper alternatives.",

    "The AirPods are fine. They're easy to set up and use, but the sound lacks the clarity I was hoping for. Average at best.",

    "They do what they claim to do. The connection is stable, battery is decent, but I'm not wowed by the performance. Just adequate.",

    "I'm undecided about these. They're convenient for the Apple ecosystem, but the sound quality doesn't justify the premium pricing.",

    "The AirPods are acceptable for daily use. Comfort is good, but the audio quality is nothing special. I've heard better for less money.",

    "They're okay earbuds. Battery life is standard, sound is decent, and they fit fine. I just expected more innovation for the price.",

    "These work fine for basic listening. The connection is reliable, but the sound quality is just average. Not bad, but not great either.",

    "The AirPods are alright for casual use. They're comfortable and connect easily, but the audio doesn't stand out from the competition.",

    "They're decent but unexceptional. The battery lasts okay, the sound is acceptable, but there's nothing that makes them special.",

    "I'm neutral on these. They work as advertised, but the sound quality and features don't feel premium despite the price tag.",

    "The AirPods are okay for iPhone users. The integration is good, but the sound quality is mediocre compared to dedicated audio brands.",

    "They're fine for everyday tasks. Calls sound clear enough, and music is acceptable. I just don't find them worth the investment.",

    "These are average wireless earbuds. They connect well and are comfortable, but the audio quality is nothing to get excited about.",

    "The AirPods work as expected. Battery life is okay, sound is decent, but I don't see what makes them worth the premium price.",

    "They're good for the convenience factor, but the sound quality is just average. I've had similar experiences with cheaper earbuds.",

    "I'm moderately satisfied. They're comfortable and easy to use, but the audio performance doesn't match the cost. Okay but overpriced.",

    "The AirPods are fine for Apple users. The pairing is seamless, but the sound quality is average. I expected more for this investment.",

    "They're decent earbuds. The battery is okay, the fit is comfortable, but the sound lacks the depth I was hoping for at this price.",

    "These work well enough. Connection is stable, battery is acceptable, but the audio quality doesn't justify the premium pricing.",

    "The AirPods are okay. They're convenient for Apple devices, but the sound is just average. I don't think they're worth full price.",

    "They're fine for general use. Comfortable fit, easy pairing, but the audio quality is nothing exceptional. Average overall.",

    "I'm not particularly impressed. They work fine, but the sound quality and battery life are standard. Nothing stands out.",

    "The AirPods are acceptable. They do their job, but I don't find them particularly impressive. Average performance for a high price.",

    "They're okay for everyday listening. The convenience is nice, but the sound quality doesn't match cheaper alternatives I've tried.",

    "These are decent but not remarkable. They connect easily and are comfortable, but the audio is just okay for the price point.",

    "The AirPods work fine for basic needs. Sound is acceptable, battery is decent, but I don't think they offer good value for money.",

    "They're alright. The setup is easy, and they're comfortable, but the sound quality is average at best. Nothing special.",

    "I'm neither happy nor unhappy with these. They work as advertised, but there's nothing that makes them stand out from competitors.",

    "The AirPods are fine for casual use. Battery life is okay, sound is decent, but I expected better quality for this price.",

    "They're decent wireless earbuds. Easy to use and comfortable, but the audio quality doesn't impress me. Average overall.",

    "These work adequately. The connection is stable, and they're comfortable, but the sound is just okay. Not worth the premium in my view.",

    "The AirPods are acceptable for daily use. They connect easily and fit well, but the sound quality is mediocre for the cost.",

    "They're okay. Convenient for Apple users, but the audio quality is standard. I don't see what justifies the high price tag.",

    "I'm lukewarm about these. They work fine, but the sound quality and features don't feel special. Average product at a premium price.",

    "The AirPods are fine. They're easy to use and comfortable, but the sound doesn't impress me. Adequate but not outstanding.",

    "They're decent for what they are. Battery is okay, sound is acceptable, but I've had similar experiences with much cheaper options.",

    "These work as expected. Connection is good, fit is comfortable, but the audio quality is just average. Nothing remarkable.",

    "The AirPods are okay for Apple ecosystem users. Pairing is seamless, but the sound quality is standard. Not worth the premium to me.",

    "They're fine for general listening. Comfortable and easy to use, but the audio performance is average. I expected more for this price.",

    "I'm neutral about these AirPods. They work well enough, but the sound quality and battery life are just okay. Nothing special.",

    "The AirPods are acceptable. They connect easily and are fairly comfortable, but the audio quality doesn't justify the cost.",

    "They're okay wireless earbuds. The convenience is nice, but the sound quality is average compared to the competition.",

    "These work fine for everyday use. Battery is decent, sound is okay, but I don't think they're worth the investment.",

    "The AirPods are alright. Easy to set up and use, but the audio quality is nothing special. Average at this price point.",

    "They're decent but not impressive. Comfortable fit, stable connection, but the sound quality is just okay. Nothing stands out.",

    "I'm undecided about these. They work as advertised, but the performance doesn't feel premium despite the price tag.",

    # ========================= NEGATIVE REVIEWS (100) =========================
    "I regret buying the Apple AirPods. They started disconnecting from my iPhone after just a few weeks, and the sound quality was disappointing for such an expensive product.",

    "Very disappointed with my AirPods. They don't stay in my ears during exercise, and the battery drains too quickly. I was expecting much more for the price.",

    "The AirPods are overrated. The sound is mediocre, and they don't block out any noise despite claiming to have noise cancellation. I wouldn't buy them again.",

    "Terrible experience. The connection drops constantly, and the charging case stopped working within a month. Not worth the money at all.",

    "I had high hopes, but these AirPods are a letdown. The sound is tinny, they hurt my ears, and the battery life is shorter than advertised.",

    "Complete waste of money! The left AirPod stopped working after two months, and customer service was unhelpful. Very frustrating experience.",

    "These are awful. They fall out constantly, the connection is unreliable, and the sound quality is poor. I regret this purchase completely.",

    "Very disappointed! The battery life is terrible‚Äîthey barely last 2 hours on a full charge. The sound is also underwhelming for the price.",

    "Don't waste your money on these! They disconnect randomly, the fit is uncomfortable, and the charging case is flimsy. Terrible product.",

    "I'm extremely unhappy with my AirPods. The connection drops every few minutes, and the sound quality is worse than my old $20 earbuds.",

    "These are overpriced junk! The left AirPod died within weeks, and the sound is muffled and tinny. Apple should be ashamed of this product.",

    "Horrible experience! They won't stay in my ears, the battery drains in an hour, and they keep disconnecting from my phone. Total garbage.",

    "I'm very disappointed. The AirPods are uncomfortable, the sound quality is poor, and they stopped charging after a month. Waste of money!",

    "These are the worst earbuds I've ever owned. The connection is terrible, they fall out constantly, and the battery life is pathetic.",

    "Not worth the hype! The sound is mediocre, they don't fit properly, and the battery dies way too quickly. I expected much better from Apple.",

    "Terrible product! The left AirPod stopped working completely after three weeks. The sound was never great to begin with. Very disappointed.",

    "I hate these AirPods! They're uncomfortable, the connection is spotty, and the battery life is awful. I've wasted my money on this trash.",

    "Very poor quality! The charging case broke within two weeks, and the AirPods themselves have terrible sound. I want my money back!",

    "These are garbage! They disconnect constantly, the sound is tinny, and they hurt my ears after 10 minutes. Absolutely terrible purchase.",

    "I'm furious about this purchase! The AirPods stopped charging after a month, and the sound quality was never good. Complete rip-off!",

    "Worst earbuds ever! They fall out during any movement, the battery dies in less than an hour, and the connection drops constantly. Horrible!",

    "Don't buy these! The left AirPod failed after just a few weeks, and the right one sounds terrible. Apple quality has really gone downhill.",

    "These are absolutely terrible! The fit is uncomfortable, the sound is poor, and they disconnect from my phone every few minutes. Waste of money!",

    "I'm extremely disappointed! The battery life is nowhere near advertised, and the sound quality is worse than my cheap wired earbuds.",

    "Horrible product! The AirPods won't stay in my ears, the connection is unreliable, and the charging case stopped working after two weeks.",

    "These are a complete disaster! The sound is awful, they're uncomfortable, and the left one died after a month. I'm so angry about this purchase!",

    "Don't waste your money! The connection drops constantly, the battery drains in minutes, and the sound quality is terrible. Absolutely awful!",

    "I regret buying these! They're uncomfortable, the sound is tinny and flat, and they disconnect from my devices constantly. Very disappointed!",

    "Terrible quality! The charging case broke within weeks, and the AirPods themselves have poor sound and keep falling out. Complete waste!",

    "These are the worst! They don't fit, they don't sound good, and the battery life is pathetic. I expected so much better from Apple.",

    "Awful experience! The left AirPod stopped working after a month, and the sound was never impressive. Customer service was no help either.",

    "I'm so disappointed! These AirPods are uncomfortable, the connection is terrible, and the battery dies way too fast. Not worth it at all!",

    "Complete garbage! They fall out constantly, the sound is mediocre, and the battery life is much shorter than advertised. Don't buy!",

    "These are terrible! The connection drops every few minutes, they hurt my ears, and the sound quality is worse than my old earbuds.",

    "Very poor product! The AirPods stopped charging properly after two weeks, and the sound was never good. I want a refund!",

    "I hate these! They're uncomfortable, the battery dies quickly, and they disconnect constantly. Worst purchase I've made in years!",

    "Don't waste your hard-earned money! The sound is awful, they don't fit right, and the left one died within a month. Terrible!",

    "These are completely worthless! The connection is unreliable, the battery life is pathetic, and the sound quality is disappointing.",

    "Awful earbuds! They fall out during exercise, the sound is tinny, and the charging case is poorly designed. Very unhappy with this!",

    "I'm extremely frustrated! The AirPods disconnect randomly, the battery drains in an hour, and the sound quality is terrible for the price.",

    "These are a joke! The left AirPod failed after three weeks, and the sound was never impressive. Apple really dropped the ball here.",

    "Horrible investment! They're uncomfortable, the connection is spotty, and the battery life is way shorter than promised. Big disappointment!",

    "Don't buy these AirPods! The sound is poor, they don't stay in my ears, and the charging case stopped working after a few weeks.",

    "I'm so angry about this purchase! The AirPods are uncomfortable, the battery dies fast, and they disconnect constantly. Total waste!",

    "These are the worst earbuds I've tried! The sound is muffled, they fall out easily, and the battery life is nowhere near advertised.",

    "Terrible experience! The left AirPod died after a month, the sound is mediocre, and the fit is uncomfortable. I regret buying these!",

    "Complete disappointment! The connection drops constantly, the battery drains too quickly, and the sound quality is poor. Not recommended!",

    "These are awful! They hurt my ears, the sound is tinny, and they disconnect from my phone every few minutes. Horrible product!",

    "I'm very unhappy! The AirPods won't hold a charge, the sound is underwhelming, and they fall out constantly. Waste of money!",

    "Don't believe the hype! These AirPods are uncomfortable, have terrible battery life, and the sound quality is disappointing. Not worth it!",

    "Terrible build quality! The charging case feels cheap, the AirPods rattle inside, and the sound is poor. Very disappointed with Apple!",

    "These are useless! The battery dies within an hour, they disconnect constantly, and the sound quality is awful. Complete waste of money!",

    "I'm so frustrated! The AirPods won't pair properly, the battery life is terrible, and they're uncomfortable to wear. Horrible product!",

    "Worst purchase ever! The left AirPod stopped working in two weeks, and the right one has terrible sound. Don't waste your money!",

    "These are trash! They don't fit my ears, the sound is poor, and the battery drains incredibly fast. I'm so disappointed!",

    "Horrible quality! The connection is unstable, the sound is tinny, and they hurt my ears after wearing them for just minutes.",

    "I regret this purchase so much! The AirPods are uncomfortable, the battery life is awful, and the sound quality is disappointing.",

    "These are terrible earbuds! They fall out constantly, the connection drops frequently, and the sound is worse than cheap alternatives.",

    "Don't buy these! The charging case stopped working after a month, and the sound quality was never good. Complete disappointment!",

    "I'm extremely unhappy! The battery drains too fast, the fit is uncomfortable, and the sound quality is terrible for this price.",

    "Awful product! The AirPods disconnect randomly, the battery life is pathetic, and the sound is muffled and unclear. Not worth it!",

    "These are garbage! The left AirPod died within weeks, the sound is poor, and the charging case is flimsy. Terrible quality!",

    "I'm so disappointed with these! They're uncomfortable, the battery dies quickly, and the connection is unreliable. Complete waste!",

    "Horrible experience! The sound quality is terrible, they won't stay in my ears, and the battery life is much shorter than advertised.",

    "These are the worst! The connection drops constantly, they're uncomfortable, and the sound quality is disappointing. Don't buy!",

    "I hate these AirPods! The battery drains in an hour, they fall out easily, and the sound is tinny. Terrible investment!",

    "Complete trash! The charging case broke after two weeks, the sound is poor, and they disconnect from my phone constantly.",

    "I'm furious! The AirPods stopped working after a month, the sound was always poor, and customer service was unhelpful. Awful!",

    "These are worthless! The fit is terrible, the sound quality is poor, and the battery life is nowhere near what Apple claims.",

    "Don't waste your money! The left AirPod failed within weeks, the sound is mediocre, and they're uncomfortable to wear. Terrible!",

    "Horrible product! They disconnect every few minutes, the battery dies too quickly, and the sound quality is disappointing.",

    "I'm very disappointed! The AirPods are uncomfortable, the connection is unreliable, and the battery life is pathetic. Not worth it!",

    "These are awful! The sound is tinny, they hurt my ears, and the charging case stopped working after a month. Complete waste!",

    "Terrible quality! The AirPods won't stay in my ears, the battery drains fast, and the sound is worse than cheap earbuds.",

    "I regret buying these! The connection drops constantly, the fit is uncomfortable, and the sound quality is terrible for the price.",

    "These are garbage! The left AirPod died after two weeks, the sound is poor, and the battery life is awful. Don't buy!",

    "Horrible experience! They're uncomfortable, the connection is spotty, and the battery drains in less than an hour. Complete disappointment!",

    "I'm so unhappy! The AirPods fall out constantly, the sound is mediocre, and the charging case broke within a month. Terrible!",

    "These are the worst earbuds I've owned! The battery life is pathetic, the sound is tinny, and they disconnect all the time.",

    "Don't buy these AirPods! The quality is terrible, they're uncomfortable, and the battery dies way too quickly. Complete waste!",

    "I'm extremely frustrated! The sound quality is poor, they won't stay in my ears, and the connection drops constantly. Awful!",

    "These are useless! The left AirPod stopped working after three weeks, and the sound was always disappointing. Not worth it!",

    "Horrible product! The battery drains too fast, the fit is uncomfortable, and the sound quality doesn't justify the price at all.",

    "I hate these! They disconnect constantly, the sound is terrible, and the charging case feels cheap. Worst purchase ever!",

    "Complete disaster! The AirPods are uncomfortable, the battery life is awful, and the sound quality is disappointing. Don't buy!",

    "These are trash! The connection is unreliable, they fall out easily, and the sound is worse than my old wired earbuds.",

    "I'm so disappointed! The battery dies in an hour, the fit is terrible, and the sound quality is poor. Complete waste of money!",

    "Awful quality! The left AirPod failed within a month, the sound is tinny, and they're uncomfortable to wear. Not recommended!",

    "Don't waste your money on these! The connection drops constantly, the battery life is pathetic, and the sound is terrible.",

    "I regret this purchase! The AirPods are uncomfortable, the sound quality is poor, and the battery drains way too quickly.",

    "These are horrible! They won't stay in my ears, the connection is spotty, and the sound quality is disappointing for this price.",

    "Terrible product! The charging case broke after two weeks, the battery life is awful, and the sound is mediocre at best.",

    "I'm very unhappy! The AirPods disconnect randomly, the fit is uncomfortable, and the sound quality is terrible. Don't buy!",

    "These are the worst! The battery drains fast, they fall out constantly, and the sound is tinny and poor. Complete waste!",

    "Horrible investment! The left AirPod died after a month, the sound is disappointing, and the connection is unreliable. Awful!",

    "Don't buy these AirPods! The quality is poor, they're uncomfortable, and the battery life is much shorter than advertised.",

    "I'm extremely disappointed! The sound is terrible, they won't stay in my ears, and the charging case stopped working quickly.",

    "These are garbage! The connection drops constantly, the battery life is pathetic, and the sound quality is awful. Not worth it!",

    "Terrible experience! The AirPods are uncomfortable, the sound is poor, and the battery drains way too fast. Complete disappointment!",

    "I hate these! They fall out during exercise, the connection is unreliable, and the sound quality is terrible for the price.",

    "Awful product! The left AirPod failed within weeks, the sound is mediocre, and the battery life is disappointing. Don't buy!"
]


actual_sentiments = ["Positive"] * 100 + ["Neutral"] * 100 + ["Negative"] * 100


# Verification
print(f"Total reviews: {len(manual_reviews)}")
print(f"Total sentiments: {len(actual_sentiments)}")
print(f"Positive: {actual_sentiments.count('Positive')}")
print(f"Neutral: {actual_sentiments.count('Neutral')}")
print(f"Negative: {actual_sentiments.count('Negative')}")

# ============================================
# PREDICTION FUNCTION
# ============================================

def predict_sentiment(texts, model, tokenizer, device, batch_size=8):
    model.eval()
    all_predictions = []
    all_confidences = []
    label_map = {0: 'Negative', 1: 'Neutral', 2: 'Positive'}

    for i in range(0, len(texts), batch_size):
        batch_texts = texts[i:i+batch_size]

        encodings = tokenizer(
            batch_texts,
            truncation=True,
            padding=True,
            max_length=512,
            return_tensors='pt'
        )

        input_ids = encodings['input_ids'].to(device)
        attention_mask = encodings['attention_mask'].to(device)

        with torch.no_grad():
            outputs = model(input_ids=input_ids, attention_mask=attention_mask)
            logits = outputs.logits
            probabilities = torch.nn.functional.softmax(logits, dim=-1)
            predictions = torch.argmax(logits, dim=-1)
            confidences = torch.max(probabilities, dim=-1).values

            all_predictions.extend(predictions.cpu().numpy())
            all_confidences.extend(confidences.cpu().numpy())

    predicted_labels = [label_map[pred] for pred in all_predictions]
    return all_predictions, predicted_labels, all_confidences

# ============================================
# RUN PREDICTIONS
# ============================================

print("\nRunning predictions...\n")
pred_indices, predicted_sentiments, confidence_scores = predict_sentiment(
    manual_reviews, model, tokenizer, device
)

# ============================================
# DISPLAY RESULTS
# ============================================

for i, (review, actual, predicted, confidence) in enumerate(zip(manual_reviews, actual_sentiments, predicted_sentiments, confidence_scores)):
    is_correct = "‚úÖ" if predicted == actual else "‚ùå"
    print(f"\nReview {i+1}: {review}")
    print(f"Actual Sentiment: {actual}")
    print(f"Predicted Sentiment: {predicted} {is_correct}")
    print(f"Confidence Score: {confidence*100:.2f}%")
    print("-" * 50)
```

```python
# ============================================
# CALCULATE AND DISPLAY PER-CLASS ACCURACY & REPORT
# ============================================

label_to_num = {'Negative': 0, 'Neutral': 1, 'Positive': 2}
actual_labels_num = [label_to_num[s] for s in actual_sentiments]
predicted_labels_num = list(pred_indices)

# Per-class accuracy (calculated from confusion matrix)
cm = confusion_matrix(actual_labels_num, predicted_labels_num)
per_class_accuracy = cm.diagonal() / cm.sum(axis=1)

print(f"{'='*50}")
print("PER-CLASS ACCURACY")
print(f"{'='*50}")
print(f"Negative Class Accuracy: {per_class_accuracy[0]*100:.2f}%")
print(f"Neutral Class Accuracy: {per_class_accuracy[1]*100:.2f}%")
print(f"Positive Class Accuracy: {per_class_accuracy[2]*100:.2f}%")

# Detailed classification report
print(f"\n{'='*50}")
print("DETAILED CLASSIFICATION REPORT")
print(f"{'='*50}")
print(classification_report(
    actual_labels_num,
    predicted_labels_num,
    target_names=['Negative', 'Neutral', 'Positive'],
    digits=4
))
```
